```python
import numpy as np
import matplotlib.pyplot as plt

# Define the activation functions and their derivatives
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)

def tanh(x):
    return np.tanh(x)

def tanh_derivative(x):
    return 1 - x**2

def relu(x):
    return np.maximum  (0, x)

def relu_derivative(x):
    return (x > 0).astype(int)

# Initialize weights and biases
def initialize_weights(input_size, hidden_size, output_size, initialization='random'):
    if initialization == 'random':
        return np.random.randn(input_size, hidden_size), np.random.randn(hidden_size, output_size)
    elif initialization == 'he':
        return np.random.randn(input_size, hidden_size) * np.sqrt(2.0 / input_size), np.random.randn(hidden_size, output_size) * np.sqrt(2.0 / hidden_size)
    elif initialization == 'xavier':
        d = np.sqrt(6.0 / (input_size + output_size))
        return np.random.uniform(-d, d, (input_size, hidden_size)), np.random.uniform(-d, d, (hidden_size, output_size))

# Define the neural network architecture
input_size = 1
hidden_size = 10
output_size = 1
learning_rate = 0.01
num_epochs = 1000

# Generate some random training data
np.random.seed(0)
X = np.random.rand(100, input_size)
y = 3 * X + np.random.randn(100, output_size) * 0.1

# Initialize weights and biases for different initialization techniques
initializations = ['random', 'he', 'xavier']
activation_functions = ['relu', 'sigmoid', 'tanh']

# Create subplots for comparison
fig, axes = plt.subplots(len(initializations), len(activation_functions), figsize=(15, 12))

for i, initialization in enumerate(initializations):
    for j, activation_function in enumerate(activation_functions):
        # Initialize weights and biases with the selected technique
        weights_input_hidden, weights_hidden_output = initialize_weights(input_size, hidden_size, output_size, initialization)
        if initialization == 'random':
            # For random initialization, we set the initial values to 0
            bias_hidden = np.zeros((1, hidden_size))
            bias_output = np.zeros((1, output_size))
        else:
            # For He and Xavier, bias can be initialized as zero
            bias_hidden = np.zeros((1, hidden_size))
            bias_output = np.zeros((1, output_size))

        # Lists to store loss values during training
        losses = []

        # Training loop
        for epoch in range(num_epochs):
            # Forward propagation
            hidden_input = np.dot(X, weights_input_hidden) + bias_hidden
            if activation_function == 'relu':
                hidden_output = relu(hidden_input)
            elif activation_function == 'sigmoid':
                hidden_output = sigmoid(hidden_input)
            elif activation_function == 'tanh':
                hidden_output = tanh(hidden_input)

            output = np.dot(hidden_output, weights_hidden_output) + bias_output

            # Calculate the loss (mean squared error)
            loss = np.mean((y - output) ** 2)
            losses.append(loss)

            # Backpropagation
            d_output = output - y
            d_hidden = d_output.dot(weights_hidden_output.T)

            if activation_function == 'relu':
                d_hidden *= relu_derivative(hidden_output)
            elif activation_function == 'sigmoid':
                d_hidden *= sigmoid_derivative(hidden_output)
            elif activation_function == 'tanh':
                d_hidden *= tanh_derivative(hidden_output)

            # Update weights and biases
            weights_hidden_output -= hidden_output.T.dot(d_output) * learning_rate
            weights_input_hidden -= X.T.dot(d_hidden) * learning_rate
            bias_output -= np.sum(d_output, axis=0, keepdims=True) * learning_rate
            bias_hidden -= np.sum(d_hidden, axis=0, keepdims=True) * learning_rate

        # Plot the loss curve
        axes[i, j].plot(range(num_epochs), losses)
        axes[i, j].set_xlabel('Epoch')
        axes[i, j].set_ylabel('Loss')
        axes[i, j].set_title(f'{initialization.capitalize()} Initialization, {activation_function.capitalize()} Activation')

# Adjust the layout and show the comparison plots
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the activation functions and their derivatives
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)

def tanh(x):
    return np.tanh(x)

def tanh_derivative(x):
    return 1 - x**2

def relu(x):
    return np.maximum(0, x)

def relu_derivative(x):
    return (x > 0).astype(int)

# Initialize weights and biases
def initialize_weights(input_size, hidden_size, output_size, initialization='xavier'):
    if initialization == 'random':
        return np.random.randn(input_size, hidden_size), np.random.randn(hidden_size, output_size)
    elif initialization == 'xavier':
        d = np.sqrt(6.0 / (input_size + output_size))
        return np.random.uniform(-d, d, (input_size, hidden_size)), np.random.uniform(-d, d, (hidden_size, output_size))

# Define the neural network architecture
input_size = 1
hidden_size = 10
output_size = 1
learning_rate = 0.01
num_epochs = 1000

# Generate some random training data
np.random.seed(0)
X = np.random.rand(100, input_size)
y = 3 * X + np.random.randn(100, output_size) * 0.1

# Initialize weights and biases with Xavier initialization
weights_input_hidden, weights_hidden_output = initialize_weights(input_size, hidden_size, output_size, initialization='xavier')
bias_hidden = np.zeros((1, hidden_size))
bias_output = np.zeros((1, output_size))

# Lists to store loss values during training
losses = []

# Training loop
for epoch in range(num_epochs):
    # Forward propagation
    hidden_input = np.dot(X, weights_input_hidden) + bias_hidden
    hidden_output = relu(hidden_input)
    output = np.dot(hidden_output, weights_hidden_output) + bias_output

    # Calculate the loss (mean squared error)
    loss = np.mean((y - output) ** 2)
    losses.append(loss)

    # Backpropagation
    d_output = output - y
    d_hidden = d_output.dot(weights_hidden_output.T)
    d_hidden *= relu_derivative(hidden_output)

    # Gradient Clipping
    max_gradient = 5.0  # Set your desired maximum gradient value
    d_output = np.clip(d_output, -max_gradient, max_gradient)
    d_hidden = np.clip(d_hidden, -max_gradient, max_gradient)

    # Update weights and biases
    weights_hidden_output -= hidden_output.T.dot(d_output) * learning_rate
    weights_input_hidden -= X.T.dot(d_hidden) * learning_rate
    bias_output -= np.sum(d_output, axis=0, keepdims=True) * learning_rate
    bias_hidden -= np.sum(d_hidden, axis=0, keepdims=True) * learning_rate

# Plot the loss curve
plt.plot(range(num_epochs), losses)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Loss Curve (Xavier Initialization, ReLU Activation)')
plt.show()
```

```python
import numpy as np

class BatchNormalization:
    def __init__(self, input_shape, epsilon=1e-5):
        self.epsilon = epsilon
        self.gamma = np.ones(input_shape)
        self.beta = np.zeros(input_shape)
        self.mean = None
        self.variance = None
        self.batch_size = None
        self.x_hat = None

    def forward(self, x, training=True):
        if training:
            # Calculate mean and variance for the current batch
            self.mean = np.mean(x, axis=0)
            self.variance = np.var(x, axis=0)
            
            # Normalize the input
            self.x_hat = (x - self.mean) / np.sqrt(self.variance + self.epsilon)
            
            # Scale and shift using gamma and beta
            out = self.gamma * self.x_hat + self.beta
        else:
            # Use population statistics during inference
            x_hat = (x - self.mean) / np.sqrt(self.variance + self.epsilon)
            out = self.gamma * x_hat + self.beta
        
        return out

    def backward(self, dout):
        m = dout.shape[0]
        
        dgamma = np.sum(dout * self.x_hat, axis=0)
        dbeta = np.sum(dout, axis=0)
        dx_hat = dout * self.gamma
        
        dvariance = np.sum(dx_hat * (self.epsilon - self.x_hat) * (self.x_hat ** -1.5), axis=0) * -0.5
        dmean = np.sum(dx_hat * -1 / np.sqrt(self.variance + self.epsilon), axis=0) + dvariance * np.sum(-2 * self.x_hat, axis=0) / m
        
        dx = dx_hat * 1 / np.sqrt(self.variance + self.epsilon) + dvariance * 2 * self.x_hat / m + dmean / m
        
        self.gamma -= dgamma
        self.beta -= dbeta
        
        return dx
```
