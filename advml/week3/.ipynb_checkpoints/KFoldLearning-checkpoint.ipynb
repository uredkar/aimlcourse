```python
from sklearn.model_selection import KFold
import numpy as np
from scipy import stats

# Sample data (you should replace this with your own dataset)
X = np.array([[1, 2], [2, 3], [3, 4], [5, 6], [6, 7]])
y = np.array([0, 0, 1, 1, 1])  # Binary labels (0 or 1)

# Number of folds (k)
k = 5

# Confidence level
confidence_level = 0.95

# Create a KFold object
kf = KFold(n_splits=k, shuffle=True, random_state=42)

# Initialize a variable to store the cross-validation results (e.g., accuracy)
cross_val_results = []

# Iterate through the folds
for train_idx, test_idx in kf.split(X):
    X_train, X_test = X[train_idx], X[test_idx]
    y_train, y_test = y[train_idx], y[test_idx]
    
    # Replace this with your machine learning model and evaluation metric
    # In this example, we're using a simple Logistic Regression model.
    from sklearn.linear_model import LogisticRegression
    from sklearn.metrics import accuracy_score
    
    model = LogisticRegression()
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    # Calculate and store the evaluation metric (accuracy in this case)
    accuracy = accuracy_score(y_test, y_pred)
    cross_val_results.append(accuracy)

# Calculate the mean and standard deviation of the evaluation metric
mean_accuracy = np.mean(cross_val_results)
std_accuracy = np.std(cross_val_results)

# Calculate confidence interval
n = len(cross_val_results)
margin_of_error = stats.t.ppf((1 + confidence_level) / 2, n - 1) * (std_accuracy / np.sqrt(n))
confidence_interval = (mean_accuracy - margin_of_error, mean_accuracy + margin_of_error)

print(f"Cross-Validation Results (Accuracy): {cross_val_results}")
print(f"Mean Accuracy: {mean_accuracy:.4f}")
print(f"Standard Deviation: {std_accuracy:.4f}")
print(f"Confidence Interval ({confidence_level * 100}%): {confidence_interval}")
```

```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_classification
from imblearn.over_sampling import RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler

# Generate a synthetic imbalanced dataset
X, y = make_classification(n_samples=1000, weights=[0.9, 0.1], random_state=42)

# Check the class distribution
print("Class Distribution (Before Resampling):")
print(pd.Series(y).value_counts())

# Upsample the minority class using RandomOverSampler
ros = RandomOverSampler(sampling_strategy='minority', random_state=42)
X_upsampled, y_upsampled = ros.fit_resample(X, y)

# Check the class distribution after upsampling
print("\nClass Distribution (After Upsampling):")
print(pd.Series(y_upsampled).value_counts())

# Downsample the majority class using RandomUnderSampler
rus = RandomUnderSampler(sampling_strategy='majority', random_state=42)
X_downsampled, y_downsampled = rus.fit_resample(X, y)

# Check the class distribution after downsampling
print("\nClass Distribution (After Downsampling):")
print(pd.Series(y_downsampled).value_counts())
```

```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_classification
from imblearn.over_sampling import SMOTE

# Generate a synthetic imbalanced dataset
X, y = make_classification(n_samples=1000, weights=[0.9, 0.1], random_state=42)

# Check the class distribution
print("Class Distribution (Before SMOTE):")
print(pd.Series(y).value_counts())

# Apply SMOTE to oversample the minority class
smote = SMOTE(sampling_strategy='auto', random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

# Check the class distribution after SMOTE
print("\nClass Distribution (After SMOTE):")
print(pd.Series(y_resampled).value_counts())
```

```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_classification
from imblearn.under_sampling import TomekLinks

# Generate a synthetic imbalanced dataset
X, y = make_classification(n_samples=1000, weights=[0.9, 0.1], random_state=42)

# Check the class distribution
print("Class Distribution (Before Tomek Links):")
print(pd.Series(y).value_counts())

# Apply Tomek Links to undersample the majority class
tomek_links = TomekLinks(sampling_strategy='auto')
X_resampled, y_resampled = tomek_links.fit_resample(X, y)

# Check the class distribution after Tomek Links
print("\nClass Distribution (After Tomek Links):")
print(pd.Series(y_resampled).value_counts())
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler

# Create a synthetic dataset with two features and a balanced class distribution
#X, y = make_classification(n_samples=300, n_features=2, random_state=42)
X, y = make_classification(1000, 10, random_state=42)
# Check the class distribution before resampling
print("Class Distribution (Before Resampling):")
print(pd.Series(y).value_counts())

# Apply SMOTE to oversample the minority class
smote = SMOTE(sampling_strategy='auto', random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

# Apply RandomUnderSampler to undersample the majority class
rus = RandomUnderSampler(sampling_strategy='majority', random_state=42)
X_resampled_under, y_resampled_under = rus.fit_resample(X, y)

# Create subplots for oversampling and undersampling
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Plot the original balanced data
axes[0].scatter(X[y == 0][:, 0], X[y == 0][:, 1], label='Class 0', alpha=0.6)
axes[0].scatter(X[y == 1][:, 0], X[y == 1][:, 1], label='Class 1', alpha=0.6)
axes[0].set_title('Original Balanced Data')
axes[0].legend()

# Plot the oversampled and undersampled data
axes[1].scatter(X_resampled[y_resampled == 0][:, 0], X_resampled[y_resampled == 0][:, 1], label='Class 0 (SMOTE)', alpha=0.6)
axes[1].scatter(X_resampled[y_resampled == 1][:, 0], X_resampled[y_resampled == 1][:, 1], label='Class 1 (SMOTE)', alpha=0.6)
axes[1].scatter(X_resampled_under[y_resampled_under == 0][:, 0], X_resampled_under[y_resampled_under == 0][:, 1], label='Class 0 (Undersampling)', alpha=0.6)
axes[1].scatter(X_resampled_under[y_resampled_under == 1][:, 0], X_resampled_under[y_resampled_under == 1][:, 1], label='Class 1 (Undersampling)', alpha=0.6)
axes[1].set_title('Oversampled and Undersampled Data')
axes[1].legend()

plt.tight_layout()
plt.show()
```

```python
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV
import numpy as np

# Load the Iris dataset for illustration
iris = load_iris()
X = iris.data
y = iris.target

# Create a RandomForestClassifier
rf_classifier = RandomForestClassifier()

# Define the hyperparameter grid for random search
param_dist = {
    'n_estimators': np.arange(50, 201, 10),
    'max_depth': np.arange(1, 11),
    'min_samples_split': np.arange(2, 11),
    'min_samples_leaf': np.arange(1, 11),
    'max_features': ['auto', 'sqrt', 'log2', None],
    'bootstrap': [True, False]
}

# Create a RandomizedSearchCV object
random_search = RandomizedSearchCV(
    rf_classifier, param_distributions=param_dist, n_iter=50,
    scoring='accuracy', cv=5, random_state=42, n_jobs=-1
)

# Fit the model and perform random search cross-validation
random_search.fit(X, y)

# Get the best hyperparameters and corresponding score
best_params = random_search.best_params_
best_score = random_search.best_score_

print("Randomized Search CV Results:")
print(f"Best Hyperparameters: {best_params}")
print(f"Best Accuracy Score: {best_score:.4f}")
```
